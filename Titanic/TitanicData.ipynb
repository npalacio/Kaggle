```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Gender of survivors
maleAgesSurvived = data.loc[(data['Survived'] == 1)][['Sex']]
plot = maleAgesSurvived.Sex.hist()
```

```python
# Gender of survivors
maleAgesSurvived = data.loc[(data['Survived'] == 0)][['Sex']]
plot = maleAgesSurvived.Sex.hist()
```

```python
title = 'Age of Survivors'
plot = data.loc[data['Survived'] == 1]['Age'].plot(kind='hist', title=title)
plot.set_xlabel('Age')
```

```python
title = 'Age of Deceased'
plot = data.loc[data['Survived'] == 0]['Age'].plot(kind='hist', title=title)
plot.set_xlabel('Age')
```

```python
title = 'Age of Male Survivors'
maleAgesSurvived = data.loc[(data['Sex'] == 'male') & (data['Survived'] == 1)][['Age']]
maleAgesPlot = maleAgesSurvived.plot(kind='hist',title=title)
maleAgesPlot.set_xlabel('Age')
```

```python
title = 'Age of Male Deceased'
maleAgesDeceased = data.loc[(data['Sex'] == 'male') & (data['Survived'] == 0)][['Age']]
maleAgesPlot = maleAgesDeceased.plot(kind='hist',title=title)
maleAgesPlot.set_xlabel('Age')
```

```python
title = 'Age of Female Survivors'
maleAgesSurvived = data.loc[(data['Sex'] == 'female') & (data['Survived'] == 1)][['Age']]
maleAgesPlot = maleAgesSurvived.plot(kind='hist',title=title)
maleAgesPlot.set_xlabel('Age')
```

```python
title = 'Age of Female Deceased'
maleAgesDeceased = data.loc[(data['Sex'] == 'female') & (data['Survived'] == 0)][['Age']]
maleAgesPlot = maleAgesDeceased.plot(kind='hist',title=title)
maleAgesPlot.set_xlabel('Age')
```
